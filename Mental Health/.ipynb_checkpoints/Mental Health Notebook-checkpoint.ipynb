### Abstract <a id='section-1'>

As shelter-in-place ordinances rolled into effect throughout the nation in an effort to mitigate the spread of COVID-19, the phrases “anxiety symptoms”, “worried”, and other related terms witnessed a spike as Google search queries in March 2020.

Shelter-in-place restrictions, in combination with an increased prevalence of psychological distress such as anxiety and depression during the pandemic, has highlighted an interesting question in regards to how mental health resources are being searched for.
In this investigation, we focused on quantifying Google search behavior and noting trends related to how mental health resources, particularly those provided through an online format, were sought out before and during the pandemic.

We used the Google Health Trends API to access information on relative search indexes and relative search volumes for our 3 key **search terms** : “mental health resources”, “therapy near me”, and “online therapy”. We also aimed to compare the search trends in different regions in the US based on the level of shelter-in-place restrictiveness from our WalletHub data.

Our data shows that the top two queries for *“mental health resources”* since Dec. 2019 until now are *“covid mental health resources”* which has a ***relative search index*** of 100 (the highest value) and *“free mental health resources”* with a relative search index of 64. In the same way, the *top 3 queries* for the search term *“online therapy”* were *“free therapy”*, *“online free therapy”*, and *“online therapist”* all with relative search index of 100, 100, and 60. Finally, the top two queries for the ***initial search term***, *“therapy near me”* were *“therapist”* and *“therapist near me”* with relative search indexes of 27 and 25; which are pretty low relative search index values.


Furthermore, our data displays the ***relative search volumes*** for each of the ***top queries*** associated with each of the three ***initial search terms***: "*mental health resources*", "*online therapy*", and "*therapy near me*." Overall, there seemed to be higher values outputted for the search term, "*online therapy"*.


Our preliminary assessment found that the ***most popular websites*** were *community-based* and either at the *local or national level*. The *target audience* for each website is broad and they share a multitude of online mental health resources. The types of websites most commonly visited, at positions 1 and 2, include non-profit organizations that are either local or national. These websites include but are not limited to the [National Alliance on Mental Illness](https://www.nami.org/Home), [Mental Health America](https://www.mhanational.org/), and [local county mental health resources](https://www.chesterfield.gov/878/Mental-Health-Support-Services). All of these sites target the general population, a broad audience, and provide mental health education, resources, support groups, and crisis call centers.

Our research acknowledges that searches for *negative psychological* symptoms have ***increased*** through our Google Trends exploration and literature review findings; and our preliminary findings showcase that online mental health resources are being sought during the COVID-19 pandemic and shelter-in-place policies. This brings major attention to how online search behavior can better inform public health and mental health services delivery during a time of immense psychological challenge. 


### Background  <a id='section-3'>

   The COVID-19 pandemic has introduced a number of unprecedented challenges. With shelter-in-place restrictions initiated globally, access to essential and everyday resources has been upended for many. Social and physical isolation, disrupted work and education, and concerns over health are just several stressors that have been tied with this pandemic ([McGinty](https://jamanetwork.com/journals/jama/fullarticle/10.1001/jama.2020.21231?guestAccessKey=ce3dfc95-5abc-4172-88b1-b9eb2f55a0e6&utm_source=silverchair&utm_campaign=jama_network&utm_content=covid_weekly_highlights&utm_medium=email) et al).
   
   At a global scale, a study with participants from six countries found that there were elevated psychological symptoms among those infected with COVID-19 and those that perceive themselves at risk, and that internet searches were associated with a stronger sense of internal locus among individuals during the pandemic ([Sigurvinsdottir](https://www.mdpi.com/1660-4601/17/19/6985/htm) et al). Surveys of adults in the US show that the prevalence of depression symptoms was more than three times higher during the COVID-19 pandemic than before ([Ettman](https://jamanetwork.com/journals/jamanetworkopen/fullarticle/2770146) et al). Additionally, elevated levels of loneliness has been significantly associated with greater depression and suicidal ideation since shelter-in-place policies were enacted ([Killgore](https://search-proquest-com.libproxy.berkeley.edu/psycinfo/docview/2446719807/1F05E71C21CF462BPQ/3?accountid=14496) et al.)

   This prompted us to explore online search behavior in the United States using Google Trends and to continue assessing more literature regarding the mental health implications of COVID-19 in the United States specifically. An analysis of Google search behavior has suggested that people are shifting from in-person therapy to remote therapy services ([Hoerger](https://content.apa.org/fulltext/2020-59192-001.html) et al). In combination with shelter-in-place ordinances, these findings led us to want to investigate how mental health resources were utilized and the manner in which they were searched for during the pandemic. We were particularly interested in how access to mental health resources has been affected or how their utilization has changed as a result of the shelter-in-place regulations. We wanted to expand upon previous findings and further explore the intersection of online search behavior and access to mental health resources during the coronavirus pandemic. ### Methodology  <a id='section-5'>

Some terminology that should be explained prior to getting into our methodology include:
1. relative search volume
2. relative search index
3. top queries and follow up queries/terms
4. initial search terms

A ***relative search volume*** is a normalized proportion of searches for a specific query as compared to the sum total of searches for a set of queries. A ***relative search index*** is essentially a popularity score ranging from 0-100 for search terms/queries. ***Top queries*** are queries that are most related to our initial search term and include their relative search index; whereas, ***follow up queries and terms*** are queries that follow-up after the top queries. Last, but not least, the ***initial search terms*** are the terms that we aim to explore/what we're looking for.

The steps we sought to follow were:

1. choosing an initial search term
2. finding relative search indices and explore that popularity score
3. made visualizations off of the RSI relationship
4. refer to masterlist which got rid of duplicates/redundancies


We used a similar methodology as outlined in a research paper written by Anne Zepecki and several UC Berkeley faculty on the application of Google data for health research ([Zepecki](https://www.researchprotocols.org/2020/7/e16543/) et al).


More specifically, we began our research by working on the Google Trends website and searching up general symptoms for *depression* and *anxiety*. We did this to take a stab at what our potential search terms would be. In doing so, we noted that there was a sharp uptick in searches for the terms “*anxiety symptoms*”, “*worried*”, “*fatigue*”, and “*mental health hotline*” during the month of March 2020. Notably, this was around the time many states were announcing shelter-in-place policies to help control and mitigate the spread of COVID-19.


While working with the Google Trends website, we also reviewed literature that focused on the *mental health implications* of COVID-19 on population health and recent studies found that there has been a significant increase in *psychological distress* and searches for *mental health resources*.


We then moved onto the Google API/Jupyter notebook to visualize the relationship between initial search terms and relative search index, relative search volume, and top queries. We then began creating visualizations (directed graphs and bar charts) to illustrate these relationships. Lastly, we explored the site probabilities for each initial search term.
